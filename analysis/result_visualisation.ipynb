```python
def read_numbers_from_file(filename):
    with open(filename, 'r') as file:
        numbers = [int(line.strip()) for line in file]
    return numbers

def cal_item_frequency(item_list):
    item_frequency = dict()
    for item in item_list:
        if item in item_frequency:
            item_frequency[item] += 1
        else:
            item_frequency[item] = 1
    return item_frequency

redial_0 = read_numbers_from_file('bias_analysis_data_set/selection_bias/redial_0.txt')
redial_0_IF = cal_item_frequency(redial_0)
redial_0_IF = {k: v for k, v in sorted(redial_0_IF.items(), key=lambda item: item[1], reverse=True)}
```

```python
plt.bar(range(len(redial_0_IF)), list(redial_0_IF.values()), align='center')
print(len(redial_0_IF))
plt.xlabel('Items')
plt.ylabel('Frequency')
```
